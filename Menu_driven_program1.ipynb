```python
def union(s1,s2):
    result = list(set(s1 + s2))
        
    print(f"Union of s1 and s2 is = {result}")
    
def intersection(s1,s2):
    result = []
    for i in s2:
        if i in s1:
            result.append(i)
    print(f"Intersection of s1 and s2 is = {result}")
        
def difference(s1,s2):
    result = []
    for i in s2:
        if i not in s1:
            result.append(i)
    print(f"Difference of s1 and s2 is = {result}")
        
def powerset(s):
    result = [[]]

    for  i in s:
        n = len(result)

        for j in range(n):
            r = result[j] + [i]
            result.append(r)
    for element in result:
        print(f"Powerset of s is = {element}")
            
def cartesian_product(s1,s2):
    result = [(a, b) for a in s2 for b in s1]
        
    print(f"Cartesian Product of s1 and s2 is = {result}")
    
# Creating Options
while True:
    print("SET OPERATIONS")
    print("1. Union")
    print("2. Intersection")
    print("3. Difference")
    print("4. Powerset")
    print("5. Cartesian Product")
    print("6. Exit")
    
    opt = int(input("Enter your choice "))
    
    if opt == 1:
        s1 =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        s2 =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        union(s1,s2)
    
    elif opt == 2:
        s1 =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        s2 =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        intersection(s1,s2)
        
    elif opt == 3:
        s1 =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        s2 =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        difference(s1,s2)
    
    elif opt == 4:
        s =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        powerset(s)
        
    elif opt == 5:
        s1 =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        s2 =  list(map(int,input("Enter the numbers (separated by spaces): ").split()))
        cartesian_product(s1,s2)
        
    elif opt == 6:
        break
    
    else:
        print("Choose the correct option")
    
```
